```python
# pip install gensim
import gensim
```

```python
path = "datasets/sgns.target.word-word.dynwin5.thr10.neg5.dim300.iter5.bz2"
model = gensim.models.KeyedVectors.load_word2vec_format(path, binary=False)
```

```python
x = model["你好"]
```

```python
print(len(x)) # 300
```

```python
# 王后 = （国王 - 男人）+ 女人
model.most_similar(positive=["国王", "女人"], negative=["男人"])
```

```python
words = ["房子", "车库", "商店", "狗"]
model.doesnt_match(words) # 狗
```

```python
model.similarity("男人", "女人")  # 0.87
```
